```python

# 1) List out every section header exactly as it appears
fields = [
    "Name", "Unit No", "Admission Date", "Discharge Date",
    "Date of Birth", "Sex", "Service", "Allergies", "Attending",
    "Chief Complaint", "Major Surgical or Invasive Procedure",
    "History of Present Illness", "Past Medical History",
    "Social History", "Family History", "Physical Exam",
    "Pertinent Results", "Brief Hospital Course",
    "Medications on Admission", "Discharge Medications",
    "Discharge Disposition", "Discharge Diagnosis",
    "Discharge Condition", "Discharge Instructions", "Followup Instructions"
]

# 2) Build a single regex to capture each header and its body
pattern = re.compile(
    r'(?P<field>' + '|'.join(map(re.escape, fields)) + r'):'   # one of the headers
    r'\s*(?P<value>.*?)(?=(?:' + '|'.join(map(re.escape, fields)) + r'):|$)',  # up to next header or end
    re.DOTALL
)

# 3) A helper that takes one raw note, returns a dict { header: text }
def parse_sections(text):
    matches = pattern.finditer(text)
    return {m.group("field"): m.group("value").strip() for m in matches}

# 4) Apply across the DataFrame
parsed_df = discharge_pd["text"].apply(parse_sections).apply(pd.Series)

# 5) Join back (optional)
discharge_pd = discharge_pd.join(parsed_df)

# Now you have new columns discharge_pd["Name"], discharge_pd["Unit No"], â€¦, discharge_pd["Followup Instructions"]
discharge_pd
```

```python
# 4. Build the concatenated note text (with section headers for clarity)
def build_note(row):
    sections = []
    for field in fields_to_concat:
        val = row.get(field, "")
        if pd.notnull(val) and str(val).strip():
            sections.append(f"{field}: {val}")
    return "\n".join(sections)

model_df['note_text'] = model_df.apply(build_note, axis=1)
model_df
```

```python
class ClinicalNotesDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        encoding = self.tokenizer(
            self.texts[idx],
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        encoding = {k: v.squeeze(0) for k, v in encoding.items()}
        encoding['labels'] = torch.tensor(self.labels[idx], dtype=torch.float)
        return encoding
```

```python
train_dataset = ClinicalNotesDataset(train_texts, train_labels, text_tokenizer)
val_dataset = ClinicalNotesDataset(val_texts, val_labels, text_tokenizer)
```

```python
text_model = BertForSequenceClassification.from_pretrained(
    "emilyalsentzer/Bio_ClinicalBERT",
    num_labels=len(all_codes),
    problem_type="multi_label_classification"
)
```

```python
# 7. Training arguments
training_args = TrainingArguments(
    output_dir='./clinicalbert_diagnosis',
    num_train_epochs=5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy='epoch',
    save_strategy='epoch',
    logging_strategy='steps',
    logging_steps=50,
    learning_rate=2e-5,
    warmup_steps=100,
    weight_decay=0.01,
    fp16=torch.cuda.is_available(),
    report_to="none",  # Use "wandb" or "tensorboard" if you want experiment tracking
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    save_total_limit=2,
    seed=42
)
```

```python
# 8. Compute metrics for multi-label
def compute_metrics(pred):
    logits, labels = pred
    sigmoid = torch.nn.Sigmoid()
    probs = sigmoid(torch.Tensor(logits)).cpu().numpy()
    y_pred = (probs > 0.5).astype(int)
    y_true = labels
    macro_f1 = f1_score(y_true, y_pred, average='macro', zero_division=0)
    micro_f1 = f1_score(y_true, y_pred, average='micro', zero_division=0)
    return {'macro_f1': macro_f1, 'micro_f1': micro_f1}
```

```python
# 9. Early stopping (optional but recommended for large models)
early_stop = EarlyStoppingCallback(early_stopping_patience=2)
```

```python
# 10. Trainer
trainer = Trainer(
    model=text_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
    callbacks=[early_stop]
)
```

```python
# 11. Training loop with loss/metric output per epoch
logging.info("Starting model training...")
train_result = trainer.train()
logging.info("Training complete.")
```

```python
import re

def split_sections(note):
    # Add or adjust headers as needed for your corpus
    headers = [
        "Chief Complaint", "History of Present Illness", "Past Medical History",
        "Past Surgical History", "Physical Exam", "Pertinent Results",
        "Brief Hospital Course", "Medications on Admission", "Discharge Medications",
        "Discharge Instructions"
    ]
    pattern = "(" + "|".join(re.escape(h) for h in headers) + "):"
    splits = re.split(pattern, note)
    sections = {}
    for i in range(1, len(splits), 2):
        sections[splits[i].strip()] = splits[i+1].strip()
    return sections
```
